# Classical Electronic Structure Preprocessing

This notebook demonstrates one stage of the Quantum ALD workflow. Core logic lives in `src/quantum_ald`.

```python
from pathlib import Path
import sys
sys.path.insert(0, str(Path.cwd().parent / 'src'))

from quantum_ald import H2, run_hartree_fock, run_dft, get_orbital_energies

mol = H2()
print(f'Loaded: {mol}')
mf, e_hf = run_hartree_fock(mol)
print(f'HF Energy: {e_hf:.8f} Ha')
mf_dft, e_dft = run_dft(mol, functional='lda')
print(f'DFT/LDA Energy: {e_dft:.8f} Ha')
print('Orbital energies:', get_orbital_energies(mf))
```
