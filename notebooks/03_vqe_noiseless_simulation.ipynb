# Noiseless VQE Prototype

This notebook demonstrates one stage of the Quantum ALD workflow. Core logic lives in `src/quantum_ald`.

```python
from pathlib import Path
import sys
sys.path.insert(0, str(Path.cwd().parent / 'src'))

from quantum_ald import H2, run_hartree_fock, define_active_space, map_to_qubit_hamiltonian, VQESolver

mol = H2()
mf, _ = run_hartree_fock(mol)
active_space = define_active_space(2, 2)
qubit_ham = map_to_qubit_hamiltonian(mf, active_space)
solver = VQESolver(num_qubits=4, num_electrons=2, ansatz_type='twolocal', max_iter=100)
energy, params = solver.solve(qubit_ham)
print(f'VQE ground-state energy: {energy:.8f} Ha')
```
